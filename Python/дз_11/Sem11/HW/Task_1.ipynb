```python
import numpy as np
import matplotlib.pyplot as plt
xi = np.linspace(-200,200,1000)
y = [((-12*x**4*np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30) for x in xi]
plt.plot(xi, y)

plt.grid()
plt.show()
```

 2. Интервалов на которых функция возростает и убывает бесконечное множество.

```python
# 1. Опрелелим корни. 
# Из графика видно что первый корень находиться в диапозоне (-8, -7.5)

from sympy import *
from sympy.plotting import plot

x = Symbol('x')
func = -12*x**4*sin(cos(x))-18*x**3+5*x**2+10*x-30
plot(func,(x,-8, -7.5),line_color='red')

```

из графика видно, что 1-й корень в диапозоне (-9, 9) находится в промежутке (-7.7, -7.6)
попробуем расчитать более точное значение

```python
roots = []
root = nsolve(func, -7.65)
roots.append(root)
display(root)
```

Найдем интервалы, на которых функция возрастает и убывает на рассматриваемом интервале:

найденные корни производной позволяют сделать вывод, что:

```python
display('Функция убывает на промежутках: ')
display(f'(-9;{d_roots[0]})')
display(f'({d_roots[1]};{d_roots[2]})')
display(f'({d_roots[3]};{d_roots[4]})')
display(f'({d_roots[5]};{d_roots[6]})')
```

```python
display('Функция возрастает на промежутках: ')
display(f'({d_roots[0]};{d_roots[1]})')
display(f'({d_roots[2]};{d_roots[3]})')
display(f'({d_roots[4]};{d_roots[5]})')
display(f'({d_roots[6]};9)')
```

Найдем локальные вершины на рассматриваемом диапозоне (-9, 9)

```python
def f(x):
    func = -12*x**4*sin(cos(x))-18*x**3+5*x**2+10*x-30
    return func

display('Вершины функции в диапозоне (-9, 9) имеют следующие координаты:')
for i in range (len(d_roots)):
    display(f'({d_roots[i]};{f(d_roots[i])})')
```
