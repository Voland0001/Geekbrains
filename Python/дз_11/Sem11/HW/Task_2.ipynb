Определим корни:

```python
from sympy import *
from sympy.plotting import plot

x = Symbol('x')
func = -18*x**3+5*x**2+10*x-30
plot(func,(x,-1.5, -1),line_color='red')
```

```python
root =  nsolve(func, -1.2)
display(root)
```

Найдем интервалы, на которых функция возрастает и убывает на рассматриваемом интервале: для этого найдем производную функции и корни производной.

```python
dfunc = diff(func)
display(dfunc)
```

```python
plot(dfunc,(x,-2, 2))
```

Находим корни производной:

```python
d_roots= []
d_root = nsolve(dfunc, -0.5)
d_roots.append(d_root)
display(d_root)
```

```python
d_root = nsolve(dfunc, 0.5)
d_roots.append(d_root)
display(d_root)
```

```python
display('Производная функции имеет ' + str(len(d_roots))  + ' корня:')
display(d_roots)
```

```python
display('Функция убывает на промежутках: ')
display(f'(-oo;{d_roots[0]})')
display(f'({d_roots[1]};oo)')
```

```python
display('Функция возрастает на промежутке: ')
display(f'({d_roots[0]};{d_roots[1]})')
```

У данной функции вершины нет. Найдем локальные максимальные и минимальные точки:

```python
def f(x):
    func =-18*x**3+5*x**2+10*x-30
    return func

display('Локальные min и max функции имеют следующие координаты:')
for i in range (len(d_roots)):
    display(f'({d_roots[i]};{f(d_roots[i])})')
```
